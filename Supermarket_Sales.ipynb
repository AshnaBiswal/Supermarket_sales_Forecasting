```python
#changing the datatype of date column 
df['Date']=pd.DatetimeIndex(df['Date'])
df
```

```python
#converting date colunm back into proper column
df=df.sort_values(by='Date').reset_index(drop=True)
```

```python
#plot between date and cogs column
plt.figure(figsize=(15,6))
plt.plot(df['Date'], df['cogs'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df["cogs"],freq=15)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
#displaying the trend,sesonality and noise pattern in dataset
plt.figure(figsize=(16,9))
plt.subplot(411)
plt.plot(df["cogs"], label='Original')
plt.legend(loc='upper right')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='upper right')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='upper right')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='upper right')
plt.tight_layout()
```

```python
#performing dickey fuller test 
#according to dickey fuller test 
#    (i)if p value>0.05 then we will fail to reject null hypothesis which say data is non stationary
#   (ii)if p value<0.05 then we will reject null hypothesis and accept the alternate hypothesis which say data is stationary
from statsmodels.tsa.stattools import adfuller
dftest=adfuller(df['cogs'], autolag='AIC')
dfoutput=pd.Series(dftest[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```

```python
#finding the parameters of model against the lowest AIC value
from pmdarima.arima import auto_arima
stepwise_model = auto_arima(df["cogs"], start_p=1, start_q=1,
                           max_p=3, max_q=3, m=12,
                           start_P=0, seasonal=True,
                           d=1, D=1, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
print(stepwise_model.aic())
```

```python
#importing Sarimax and passing the parameters
from statsmodels.tsa.statespace.sarimax import SARIMAX
m=SARIMAX(df['cogs'], order=(3,1,0),seasonal_order=(2,1,0,12))
res=m.fit()
```

```python
#storing the predicted values in arima_predict
df['arima_predict']=res.fittedvalues
```

```python
#plot between actual and predictes value
plt.figure(figsize=(20,12))
plt.plot(df['Date'].sort_values(),df['cogs'],label="actual")
plt.plot(df['Date'].sort_values(),df['arima_predict'],label="predicted")
plt.legend()
```

```python
#forecasting for next 15 days
forecast=res.predict(start=len(df),end=len(df)+15)
```

```python
#combining forecast with cogs 
df['cogs'].append(forecast)
```
